```python
import pandas as pd 
import re
import json
```

## Prereq-cs

```python
df = pd.read_csv('../csv/prereq-cs.csv', encoding='utf-16', delimiter='\t')
myreg = re.compile('(IC.{5})+')
split_condition = df.condition.map(lambda x :  myreg.findall(x) if myreg.findall(x) else None) 
split_course = df.course.map(lambda x :  myreg.search(x).group(0) if myreg.search(x) else None)

df2 = df
df2.condition = split_condition
df2.course = split_course
df2.cid = split_course
df2.drop_duplicates(subset="course", inplace=True)
df2 = df2.set_index('course')
df2
```

```python
df2.to_json('../json/preq-cs.json', orient='index')

# with open('preq.json', 'w') as f:
#     json.dump(f, json.loads(df2.to_json(orient='records')), indent=2)
```

## Courses-all

```python
# df = pd.read_csv('../csv/courses-all.csv', encoding='utf-16', delimiter='\t')
```

```python
all_courses = pd.read_csv('../csv/courses-all.csv', encoding='utf-16', delimiter='\t')
allc = re.compile('[^\(]*')
split_credits = all_courses.credits.map(lambda x :  allc.search(x).group(0) if allc.search(x) else None)

all_courses2 = all_courses
all_courses2.credits = split_credits
all_courses2.drop_duplicates(subset="code", inplace=True)
all_courses2 = all_courses2.set_index('code')
all_courses2
```

```python
all_courses2.to_json('../json/coures-all.json', orient='index')
```
